```python
from atomate.vasp.database import VaspCalcDb
import pandas as pd
from pymatgen import Structure
import numpy as np
import math
import os
```

```python
db = VaspCalcDb.from_db_file("db_compare_singlet_triplet.json")
col = db.collection 
```

# Se antisite

# triplet

```python
criteria = {"input.incar.NUPDOWN":2, "task_label":{"$in":["0:k:2:HSE_scf","0:k:2p:HSE_scf","0-k-HSE_scf", "0:k:HSE_scf", "0:k:1e-5:HSE_scf"]}}
sheet = []
for i in col.find(criteria):
        st = Structure.from_dict(i["output"]["structure"])
        st.to("poscar", "triplet/triplet-{}.vasp".format(i["task_id"]))
        ww_dist = st.get_distance(0,1)
        sheet.append({"task_id": i["task_id"], 
                      "nsites": i["nsites"],
                      "MAG": i["calcs_reversed"][0]["output"]["outcar"]["total_magnetization"],
#                           "date": i["calcs_reversed"][0]["completed_at"],
                      "nupdown": i["input"]["incar"]["NUPDOWN"],
                      "lz": i["input"]["structure"]["lattice"]["c"], 
                      "kpoints": i["calcs_reversed"][0]["input"]["kpoints"]["kpoints"][0],
                      "W-W": round(ww_dist, 3),
                      "Etot": i["calcs_reversed"][0]["output"]["energy"],
                      "path": i["dir_name"].split("/")[-1]
#                       "lattice": [round(st.lattice.a/math.sqrt(i["nsites"]/3),3), round(st.lattice.b/math.sqrt(i["nsites"]/3),3), round(st.lattice.c,3)]
                     })
pd.DataFrame(sheet).sort_values(["nsites", "lz"])
```

```python
relax = {"input.incar.NUPDOWN":2, 
         "task_label":{"$in":["0:k:2:p:PBE_relax","0:k:1e-5:PBE_relax","0:k:PBE_relax", "0:k:PBE_relax_FAKE", "0-k-PBE_relax"]}}
sheet = []
for i in col.find(relax):
        st = Structure.from_dict(i["output"]["structure"])
        st.to("poscar", "triplet/triplet-{}.vasp".format(i["task_id"]))
        ww_dist = st.get_distance(0,1)
        sheet.append({"task_id": i["task_id"], 
                      "nsites": i["nsites"],
                      "MAG": i["calcs_reversed"][0]["output"]["outcar"]["total_magnetization"],
#                           "date": i["calcs_reversed"][0]["completed_at"],
                      "nupdown": i["input"]["incar"]["NUPDOWN"],
                      "lz": i["input"]["structure"]["lattice"]["c"], 
                      "kpoints": i["calcs_reversed"][0]["input"]["kpoints"]["kpoints"][0],
                      "W-W": round(ww_dist, 3),
                      "Etot": i["calcs_reversed"][0]["output"]["energy"],
                      "path": i["dir_name"].split("/")[-1]
#                       "lattice": [round(st.lattice.a/math.sqrt(i["nsites"]/3),3), round(st.lattice.b/math.sqrt(i["nsites"]/3),3), round(st.lattice.c,3)]
                     })
pd.DataFrame(sheet).sort_values(["nsites", "lz"])
```
